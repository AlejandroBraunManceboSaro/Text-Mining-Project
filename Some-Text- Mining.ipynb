## Let´s check if an email is valid or not!

### Conditions:

1.format (user) @ (domain name). (Extension)\
2.letters a-z uppercase or lowercase or numbers or dots.\
3.@ and period required.\
4.The username must begin and end with alphanumerics.\
5.Username must contain 6 to 30 characters.\
6.The username cannot contain a colon in a row.\
7.Domain name from 4 to 15 alphanumeric characters.\
8.The extension must be alphabetical and longer than two characters.\

```python
import re
```

```python
mails = input("write some correct and incorrect emails spliting them by coma:")
```

```python
correct_mails=[]
incorrect_mails=[]
for mail in mails.split(","):
    if re.search('^[A-Za-z0-9]{1}([A-Za-z0-9][\.]?){4,28}[A-Za-z0-9]{1}@{1}[A-Za-z0-9]{4,15}\.{1}[a-zA-z]{2,}$',mail):
        correct_mails.append(mail)
    else:
        incorrect_mails.append(mail)
```

```python
print("This mails are correct:" + str(correct_mails))
print("This mails are incorrect:"+ str(incorrect_mails))
```

### What are the most used words by the main characters?

```python
episodeIV["episode"]="A New Hope"
episodeV["episode"]="The Empire Strikes Back"
episodeVI["episode"]="Return of the Jedi"
```

```python
episodes=pd.concat([episodeIV,episodeV,episodeVI],axis=0,ignore_index=True)
```

```python
import re
import nltk
from nltk.corpus import stopwords
import nltk as nlp
```

```python
new_column=[]
for i in episodes.dialogue:
    i=re.sub("[^a-zA-Z]", " ", i)
    i=i.lower()
    i=nltk.word_tokenize(i)
    i=[word for word in i if not word in set(stopwords.words("english"))]
    lem=nlp.WordNetLemmatizer()
    i=[lem.lemmatize(word) for word in i]
    i=" ".join(i)
    new_column.append(i)
    
```

```python
episodes["important words"]=new_column
```

## Luke

```python
luke_words=[]
frecuencia_luke=[]
for line in range(0,len(episodes[episodes.character=="LUKE"])):
    for word in episodes[episodes.character=="LUKE"]["important words"].iloc[line].split(" "):
        luke_words.append(word)
for w in luke_words:
    frecuencia_luke.append(luke_words.count(w))
```

```python
pd.DataFrame({"palabras":luke_words,"frecuencia":frecuencia_luke}).drop_duplicates().sort_values("frecuencia",ascending=False).iloc[:30]
```

## Now, Let´s create a language traductor!

```python
# Utiliza las funcionalidades de wordnet vistas en clase para crear un traductor palabra a palabra.
# El resultado debe ser la traducción al inglés de la siguiente historia
"indio querer hablar vaquero"
"vaquero disparar"
"vaquero hombre malo"
```

```python

from nltk.corpus import wordnet
```

```python
def traductor3(texto, idioma_entrada, idioma_salida):
    texto = texto.split(" ")           
    g = []
    for palabra in texto:
        for j in range(len(wordnet.synsets(palabra, lang = idioma_entrada))):
            if len(wordnet.synsets(palabra, lang = idioma_entrada)[j].lemma_names(idioma_salida))>0:
                g.append(wordnet.synsets(palabra, lang = idioma_entrada)[j].lemma_names(idioma_salida))
                break
    traducpalabra = [palabra[0] for palabra in g]
    count = 0
    for palabra in traducpalabra:
        if count==0:
            ingles = f'{palabra}'
        else:
            ingles = f'{ingles} {palabra}'
        count=1
    return(ingles)
```

```python
print(traductor3("vaquero disparar", 'spa', 'eng'))
print(traductor3("vaquero hombre malo", 'spa', 'eng'))
print(traductor3("indio querer hablar vaquero", 'spa', 'jpn'))
print(traductor3("vaquero disparar", 'spa', 'jpn'))
print(traductor3("vaquero hombre malo", 'spa', 'jpn'))
```
